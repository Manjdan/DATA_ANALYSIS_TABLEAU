# LEVEL 1: UNIVARIATE ANALYSIS
1.Separate out the categorical columns from the numerical types.
2.User defined function for calculating mean, median, mode.
3.The measure of central tendency
4.The measure of dispersion
5.visualizing the categorical columns and numerical columns

```python
# Seprating out the categorical and numerical variables
list_1 = df.select_dtypes(include="object")
list_2 = df.select_dtypes(exclude= "object")
print(tabulate({"Categorical":list_1,"numerical":list_2},headers=["Categorical","numerical"]))
```

## User-defined function for level 1 analysis of categorical columns
This function will check for the following\
Unique values\
Mode\
Total count of missing values

```python
def info_of_cat():
    for col in list_1 : 
        print(f"Unique value of {col}:{df[col].unique()}")
        print(f"mode vale of {col}:{df[col].mode()}")
        print(f"Number of missing values in {col} is {df[col].isnull().sum()}")
        print("---------------------------------")
info_of_cat()
```

## LEVEL 1 Analysis of Numerical Columns
Check if the data is discrete or continuous.\
Measure the central tendency (mean, median, mode)\
Measure the dispersion values\
Check for missing values\
Check for outliers\
Treating missing value\
Visualizing with sitable charts

```python
def info_of_numerical():
    for col in list_2:
        print(f"The mean of the {col} is {df[col].mean()}")
        print(f"The median of the {col} is {df[col].median()}")
        print(f"The mode of the {col} is {df[col].mode()[0]}")
        print(f"The standard deviation of the {col} is {df[col].std()}")
        print(f"Number of missing values in the {col} is {df[col].isnull().sum()}")
        print("--------------------------------------------------------------------\n")
info_of_numerical()
```

# OUTLIERS IN THE DATASET

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print("Outliers:")
print(outliers)
```

```python
#Visualizing the outliers with the box plot
columns = ['Avg_Open_To_Buy','Contacts_Count_12_mon','Credit_Limit','Customer_Age','Dependent_count']
df[columns].plot(kind='box', subplots=True, layout=(1, 5), figsize=(10, 5))
plt.show()
```

```python
#Visualizing the outliers with the box plot
columns = ['Months_Inactive_12_mon','Months_on_book','Total_Trans_Amt','Total_Trans_Ct']
df[columns].plot(kind='box', subplots=True, layout=(1, 4), figsize=(10, 5))
plt.show()
```

# Treating missing value

```python
# Identify missing values
missing_values = df.isnull().sum()
print("Missing values:")
print(missing_values)
```

```python
#Impute the missing values for Card category
mode_value = df["Card_Category"].mode()[0]
#Fill missing categorical object data type values with the mode value
df["Card_Category"].fillna(value=mode_value, inplace=True)
```

```python
#Impute the missing values for Credit_Limit 
mode_value = df["Credit_Limit"].mode()[0]
#Fill missing categorical object data type values with the mode value
df["Credit_Limit"].fillna(value=mode_value, inplace=True)
```
